````python
import sqlite3

# Connect to SQLite database
conn = sqlite3.connect('hospital_management.db')
cursor = conn.cursor()

# Enable foreign key support in SQLite
cursor.execute("PRAGMA foreign_keys = ON;")
conn.commit()
print("Database initialized successfully with Foreign Keys enabled.")
````

````python
# Create Tables
cursor.executescript('''
-- Departments Table
CREATE TABLE IF NOT EXISTS Departments (
    department_id INTEGER PRIMARY KEY AUTOINCREMENT,
    department_name TEXT NOT NULL UNIQUE
);

-- Doctors Table
CREATE TABLE IF NOT EXISTS Doctors (
    doctor_id INTEGER PRIMARY KEY AUTOINCREMENT,
    first_name TEXT NOT NULL,
    last_name TEXT NOT NULL,
    specialty TEXT NOT NULL,
    department_id INTEGER,
    FOREIGN KEY (department_id) REFERENCES Departments(department_id) ON DELETE SET NULL
);

-- Patients Table
CREATE TABLE IF NOT EXISTS Patients (
    patient_id INTEGER PRIMARY KEY AUTOINCREMENT,
    first_name TEXT NOT NULL,
    last_name TEXT NOT NULL,
    date_of_birth TEXT NOT NULL,
    gender TEXT CHECK(gender IN ('M', 'F', 'Other')),
    phone TEXT
);

-- Appointments Table
CREATE TABLE IF NOT EXISTS Appointments (
    appointment_id INTEGER PRIMARY KEY AUTOINCREMENT,
    patient_id INTEGER NOT NULL,
    doctor_id INTEGER NOT NULL,
    appointment_date TEXT NOT NULL, -- Format: YYYY-MM-DD
    status TEXT CHECK(status IN ('Scheduled', 'Completed', 'Cancelled')) DEFAULT 'Scheduled',
    FOREIGN KEY (patient_id) REFERENCES Patients(patient_id) ON DELETE CASCADE,
    FOREIGN KEY (doctor_id) REFERENCES Doctors(doctor_id) ON DELETE CASCADE
);

-- Treatments Table
CREATE TABLE IF NOT EXISTS Treatments (
    treatment_id INTEGER PRIMARY KEY AUTOINCREMENT,
    appointment_id INTEGER NOT NULL,
    treatment_description TEXT NOT NULL,
    cost REAL NOT NULL CHECK(cost >= 0),
    FOREIGN KEY (appointment_id) REFERENCES Appointments(appointment_id) ON DELETE CASCADE
);

-- Bills Table
CREATE TABLE IF NOT EXISTS Bills (
    bill_id INTEGER PRIMARY KEY AUTOINCREMENT,
    appointment_id INTEGER NOT NULL UNIQUE,
    total_amount REAL NOT NULL CHECK(total_amount >= 0),
    payment_status TEXT CHECK(payment_status IN ('Paid', 'Unpaid', 'Pending')) DEFAULT 'Pending',
    bill_date TEXT NOT NULL,
    FOREIGN KEY (appointment_id) REFERENCES Appointments(appointment_id) ON DELETE CASCADE
);
''')

conn.commit()
print("All tables created successfully with constraints and keys!")
````

````python
import pandas as pd

def run_query(sql_string):
    """Helper function to execute and display dataframes smoothly"""
    return pd.read_sql_query(sql_string, conn)
````

**Finding the Busiest Doctor**

````python
busiest_doctor = '''
SELECT d.doctor_id, d.first_name || ' ' || d.last_name AS doctor_name, COUNT(a.appointment_id) AS total_appointments
FROM Doctors d
JOIN Appointments a ON d.doctor_id = a.doctor_id
GROUP BY d.doctor_id
ORDER BY total_appointments DESC
LIMIT 1;
'''
run_query(busiest_doctor)
````

**Patients with more than three appointments**

````python
frequent_patients = '''
SELECT p.patient_id, p.first_name || ' ' || p.last_name AS patient_name, COUNT(a.appointment_id) AS total_appointments
FROM Patients p
JOIN Appointments a ON p.patient_id = a.patient_id
GROUP BY p.patient_id
HAVING total_appointments > 3;
'''
run_query(frequent_patients)
````

**Show today's appointments**

````python
todays_appointments = '''
SELECT a.appointment_id, p.first_name || ' ' || p.last_name AS patient_name, 
       d.first_name || ' ' || d.last_name AS doctor_name, a.appointment_date, a.status
FROM Appointments a
JOIN Patients p ON a.patient_id = p.patient_id
JOIN Doctors d ON a.doctor_id = d.doctor_id
WHERE a.appointment_date = '2026-07-08';
'''
run_query(todays_appointments)
````

**Count Appointments by Department**

````python
appointments_by_department = '''
SELECT dept.department_name, COUNT(a.appointment_id) AS total_appointments
FROM Departments dept
LEFT JOIN Doctors doc ON dept.department_id = doc.department_id
LEFT JOIN Appointments a ON doc.doctor_id = a.doctor_id
GROUP BY dept.department_name
ORDER BY total_appointments DESC;
'''
run_query(appointments_by_department)
````

**Identify Patients who have never attended an Appointment**

````python
patients_never_attended_appointment = '''
SELECT p.patient_id, p.first_name || ' ' || p.last_name AS patient_name
FROM Patients p
LEFT JOIN Appointments a ON p.patient_id = a.patient_id
WHERE a.appointment_id IS NULL;
'''
run_query(patients_never_attended_appointment)
````

**Identify doctors who have not treated any patient.**

````python
inactive_doctors = '''
SELECT d.doctor_id, d.first_name || ' ' || d.last_name AS doctor_name, d.specialty
FROM Doctors d
LEFT JOIN Appointments a ON d.doctor_id = a.doctor_id
WHERE a.appointment_id IS NULL;
'''
run_query(inactive_doctors)
````

**Find departments with below-average appointments.**

````python
low_activity_departments = '''
SELECT d.doctor_id, d.first_name || ' ' || d.last_name AS doctor_name, d.specialty
FROM Doctors d
LEFT JOIN Appointments a ON d.doctor_id = a.doctor_id
WHERE a.appointment_id IS NULL;
'''
run_query(low_activity_departments)
````

**Create a patient treatment history view.**


````python
# Execute View definition via cursor
cursor.execute('''
CREATE VIEW IF NOT EXISTS PatientTreatmentHistory AS
SELECT p.patient_id, p.first_name || ' ' || p.last_name AS patient_name,
       a.appointment_date, d.first_name || ' ' || d.last_name AS doctor_name,
       t.treatment_description, t.cost
FROM Patients p
JOIN Appointments a ON p.patient_id = a.patient_id
JOIN Doctors d ON a.doctor_id = d.doctor_id
JOIN Treatments t ON a.appointment_id = t.appointment_id;
''')
conn.commit()

# Test and show the contents of the view
run_query("SELECT * FROM PatientTreatmentHistory;")
````

**Calculate monthly revenue trends.**

````python
monthly_revenue_trends = '''
SELECT strftime('%Y-%m', bill_date) AS month, 
       SUM(total_amount) AS monthly_revenue,
       COUNT(bill_id) AS total_bills_issued
FROM Bills
WHERE payment_status = 'Paid'
GROUP BY month
ORDER BY month ASC;
'''
run_query(monthly_revenue_trends)
````

**Identify repeat patients (five or more visits).**

````python
repeat_patients = '''
SELECT p.patient_id, p.first_name || ' ' || p.last_name AS patient_name, COUNT(a.appointment_id) AS total_visits
FROM Patients p
JOIN Appointments a ON p.patient_id = a.patient_id
GROUP BY p.patient_id
HAVING total_visits >= 5;
'''
run_query(repeat_patients)
````

**Find the doctor with the highest average bill value.**

````python
doctor_highest_average_bill = '''
SELECT d.doctor_id, d.first_name || ' ' || d.last_name AS doctor_name, AVG(b.total_amount) AS average_bill_value
FROM Doctors d
JOIN Appointments a ON d.doctor_id = a.doctor_id
JOIN Bills b ON a.appointment_id = b.appointment_id
GROUP BY d.doctor_id
ORDER BY average_bill_value DESC
LIMIT 1;
'''
run_query(doctor_highest_average_bill)
````

**Testing Foreign Key Violations**

````python
# This test attempts to insert an appointment for a patient who does not exist, patient_id 999
try:
    cursor.execute('''
    INSERT INTO Appointments (patient_id, doctor_id, appointment_date, status) 
    VALUES (999, 1, '2026-07-08', 'Scheduled');
    ''')
    conn.commit()
except sqlite3.IntegrityError as e:
    print(f"{e}")
````

**Testing CHECK constraints**

````python
# This test attempts to insert a patient with an invalid gender classification ('X'). It should fail.
try:
    cursor.execute('''
    INSERT INTO Patients (first_name, last_name, date_of_birth, gender) 
    VALUES ('Invalid', 'Test', '2000-01-01', 'X');
    ''')
    conn.commit()
except sqlite3.IntegrityError as e:
    print(f"Gender CHECK Constraint caught the error: {e}")
````

**Testing ON DELETE CASCADE**

````python
# If we delete an appointment, the corresponding entries in the Treatments and Bills tables should automatically vanish.
# Testing on appointment_id = 2

# Check how many records exist for appointment 2 before deletion
print("Before Deletion:")
print("Bills count:", run_query("SELECT COUNT(*) FROM Bills WHERE appointment_id = 2").values[0][0])
print("Treatments count:", run_query("SELECT COUNT(*) FROM Treatments WHERE appointment_id = 2").values[0][0])

# Delete the main appointment record
cursor.execute("DELETE FROM Appointments WHERE appointment_id = 2;")
conn.commit()

print("\nAfter Deletion (Testing CASCADE):")
print("Bills count:", run_query("SELECT COUNT(*) FROM Bills WHERE appointment_id = 2").values[0][0])
print("Treatments count:", run_query("SELECT COUNT(*) FROM Treatments WHERE appointment_id = 2").values[0][0])
````

**ERD Design Explanation**

* **Departments ───◄ Doctors (One-to-Many)**

Relationship: One department can house many doctors; a doctor belongs to exactly one department.

Link: Doctors.department_id references Departments.department_id.

* **Patients ───◄ Appointments (One-to-Many)**

Relationship: One patient can book multiple appointments over time.

Link: Appointments.patient_id references Patients.patient_id.

* **Doctors ───◄ Appointments (One-to-Many)**

Relationship: One doctor can be assigned to multiple appointments.

Link: Appointments.doctor_id references Doctors.doctor_id.

* **Appointments ───◄ Treatments (One-to-Many)**

Relationship: An individual appointment can involve one or more specific treatments.

Link: Treatments.appointment_id references Appointments.appointment_id.

* **Appointments ─── Bills (One-to-One)**

Relationship: Each appointment generates exactly one unique medical bill.

Link: Bills.appointment_id references Appointments.appointment_id with a UNIQUE constraint.
